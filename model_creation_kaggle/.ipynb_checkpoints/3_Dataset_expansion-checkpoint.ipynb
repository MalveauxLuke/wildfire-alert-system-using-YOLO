```python
import pandas as pd
import shutil
from pathlib import Path
import random
```

```python
image_folder = Path('/kaggle/input/wildfire-smoke-detection/wildfire_smoke_dataset/wildfire_smoke_dataset/train/images')
label_folder = Path('/kaggle/input/wildfire-smoke-detection/wildfire_smoke_dataset/wildfire_smoke_dataset/train/labels')
output_image_folder = Path('/kaggle/working/images')
output_label_folder = Path('/kaggle/working/label')
sample_size = 3000
output_image_folder.mkdir(parents = True, exist_ok = True)
output_label_folder.mkdir(parents = True, exist_ok = True)
```

```python
all_images = [img for img in image_folder.iterdir()]
selected_images = random.sample(all_images, sample_size)
```

```python
for img_path in selected_images:
    shutil.copy(img_path, output_image_folder / img_path.name)
    label_path = label_folder / (img_path.stem + '.txt')
    shutil.copy(label_path, output_label_folder/ (img_path.stem + '.txt'))
```

```python
image_folder = Path('/kaggle/working/images')
val_path_images = Path('/kaggle/working/output/valid/images')
val_path_labels = Path('/kaggle/working/output/valid/labels')

train_path_images = Path('/kaggle/working/output/train/images')
train_path_labels = Path('/kaggle/working/output/train/labels')

test_path_images = Path ('/kaggle/working/output/test/images')
test_path_labels = Path ('/kaggle/working/output/test/labels')

val_path_images.mkdir(parents = True, exist_ok = True)
val_path_labels.mkdir(parents = True, exist_ok = True)

train_path_images.mkdir(parents = True, exist_ok = True)
train_path_labels.mkdir(parents = True, exist_ok = True)

test_path_images.mkdir(parents = True, exist_ok = True)
test_path_labels.mkdir(parents = True, exist_ok = True)


train_ratio = 0.7
val_ratio = 0.2
images_to_split = [img for img in image_folder.iterdir()]
total = len(images_to_split)
train_end = int(total * train_ratio)
val_end = train_end + int(total * val_ratio)
train_files = images_to_split[:train_end]
val_files = images_to_split[train_end:val_end]
test_files = images_to_split[val_end:]
```

```python
for file in train_files:
    shutil.copy(file, train_path_images / file.name)
    label_path = label_folder / (file.stem + '.txt')
    shutil.copy(label_path, train_path_labels / (file.stem + '.txt'))
for file in val_files:
    shutil.copy(file, val_path_images / file.name)
    label_path = label_folder / (file.stem + '.txt')
    shutil.copy(label_path, val_path_labels / (file.stem + '.txt'))
for file in test_files:
    shutil.copy(file, test_path_images / file.name)
    label_path = label_folder / (file.stem + '.txt')
    shutil.copy(label_path, test_path_labels / (file.stem + '.txt'))
```
